### 1.Crops Yield Data:

**The Pearson correlation analysis reveals no strong linear relationship between crop yield and the climatic or agronomic variables when considering the dataset as a whole. Yield exhibits almost no correlation with average annual rainfall (r = −0.004) and pesticide use (r = 0.066), while average temperature shows only a weak negative association (r = −0.11). These weak correlations are likely explained by the heterogeneous nature of the dataset, which combines multiple crop species, countries, and years, each characterized by distinct climatic requirements and management practices. Consequently, global linear correlations mask crop-specific relationships and suggest that more detailed analyses (e.g., stratified by crop or using nonlinear models) are required to capture the true influence of these variables on yield.**

# 3.PCA

## 1.Fit PCA on numeric variables

```python
target = yield_df["hg/ha_yield"]

numeric_features = ["average_rain_fall_mm_per_year", "pesticides_tonnes", "avg_temp"]

X_numeric = yield_df[numeric_features].copy()

# Standardize: PCA is scale-sensitive, and Rainfall_mm (hundreds) and
# Temperature_Celsius (tens) are on very different scales.
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_numeric)
```

```python
pca = PCA()
pca_scores = pca.fit_transform(X_scaled)
# Explained variance per component
explained_var = pca.explained_variance_ratio_
cumulative_var = np.cumsum(explained_var)

for i, (var, cum) in enumerate(zip(explained_var, cumulative_var), start=1):
    print(f"PC{i}: {var:.2%} explained, {cum:.2%} cumulative")
```

## 2.Visualize PCA components

```python
scree = (pca.explained_variance_ratio_*100).round(2)
scree_cum = scree.cumsum().round()
x_list = range(1, 4)
plt.bar(x_list, scree)
plt.plot(x_list, scree_cum,c="red",marker='o')
plt.xlabel("Principal Component")
plt.ylabel("Explained Variance Ratio")
plt.title("Cumulative Explained Variance")
plt.show(block=False)
```

## 3.Varaibles contribution in PCA components

```python
pca_df = pd.DataFrame(
    pca.components_.T,
    columns=[f"PC{i+1}" for i in range(pca.n_components_)],
    index=numeric_features,
)

print(pca_df.round(3))
```

```python
plt.figure(figsize=(8, 5))
sns.heatmap(pca_df, annot=True, cmap="coolwarm", center=0)
plt.title("PCA — Crop Yield Production Dataset")
plt.tight_layout()
plt.show()
```

**Key observations:**
* PC1 (45.85%) — the dominant real-world signal
 All three variables load positively and substantially: average_rain_fall_mm_per_year (0.693), avg_temp (0.610), pesticides_tonnes (0.384). This is a genuine composite factor — countries/years with more rainfall also tend to run warmer and use more pesticides, moving together. That's a coherent real-world pattern (agricultural intensification tends to cluster in warmer, wetter, more productive growing regions)

* PC2 (32.44%) — pesticide use as an independent axis
 pesticides_tonnes dominates alone (0.867), with avg_temp pulling the opposite direction (-0.497) and rainfall nearly silent (-0.043). This separates out pesticide intensity as its own dimension, somewhat decoupled from climate — likely reflecting agricultural policy/economic differences between countries rather than pure climate effects.

* PC3 (21.71%) — rainfall vs. temperature/pesticides contrast
 Rainfall (0.719) opposes both temperature (-0.617) and pesticides (-0.318) — a residual axis capturing cases where high rainfall doesn't coincide with the usual warm/high-pesticide pattern.

**Conclusion**
PCA on the real-world dataset (Crop yield prediction) shows a markedly different structure than the synthetic dataset (Crop yield agriculture): variance is unevenly concentrated (45.85% on the first component alone, vs. a flat ~20% per component in Crop yield agriculture), and rainfall, temperature, and pesticide use load together on a dominant first component — confirming genuine correlation between climate and agricultural intensity in real-world data. This contrast reinforces that Crop yield agriculture's near-uniform PCA result reflects its synthetic origin rather than a genuine absence of relationships between agronomic variables in reality
