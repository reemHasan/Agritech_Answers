# ⚙️ Segment 1: Data Integrity & Aesthetic Pre-processing

Every great analysis begins with a solid foundation. In this segment, we load our libraries, configure our visual **brand**, and perform a full multi-file audit of the **Crop Yield Prediction** dataset. We are not just cleaning data — we are preparing it for a high-dimensional agricultural intelligence pipeline.

**Objectives:**
1. Global configuration for **Plotly & Seaborn**.
2. Load and audit all 5 source files.
3. Feature Engineering for **Climate Intensity** and **Yield Efficiency** metrics.

```python
# ============================================================
# 📦 1.1 — LIBRARY IMPORTS & GLOBAL CONFIG
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
from plotly.subplots import make_subplots
from IPython.display import display, HTML
import warnings

warnings.filterwarnings('ignore')
sns.set_theme(style="darkgrid")
pio.renderers.default = "iframe"

# ── Brand Palette ──────────────────────────────────────────
C_GOLD    = "#f5a623"   # Primary accent (harvest gold)
C_BLUE    = "#4fc3f7"   # Secondary (sky / water)
C_DARK    = "#0a1628"   # Deep background
C_GREEN   = "#66bb6a"   # Growth green
C_RED     = "#ef5350"   # Alert / low yield
C_PURPLE  = "#ab47bc"   # Tertiary

PALETTE   = [C_GOLD, C_BLUE, C_GREEN, C_RED, C_PURPLE,
             "#26c6da", "#ffa726", "#ec407a", "#8d6e63", "#78909c"]

print("🚀 Agricultural Intelligence Pipeline — Environment Ready.")
```

```python
# ============================================================
# 📥 1.2 — MULTI-FILE DATA INGESTION & QUALITY AUDIT
# ============================================================

# Load all source files
df_yield       = pd.read_csv('/kaggle/input/datasets/patelris/crop-yield-prediction-dataset/yield.csv')
df_pesticides  = pd.read_csv('/kaggle/input/datasets/patelris/crop-yield-prediction-dataset/pesticides.csv')
df_rainfall    = pd.read_csv('/kaggle/input/datasets/patelris/crop-yield-prediction-dataset/rainfall.csv')
df_temp        = pd.read_csv('/kaggle/input/datasets/patelris/crop-yield-prediction-dataset/temp.csv')
df             = pd.read_csv('/kaggle/input/datasets/patelris/crop-yield-prediction-dataset/yield_df.csv')

# Rename for consistency
df.rename(columns={'Unnamed: 0': 'Record_ID'}, inplace=True)
df_rainfall.rename(columns={' Area': 'Area'}, inplace=True)

def data_audit(dataframe, name):
    audit = pd.DataFrame({
        'Dtype'         : dataframe.dtypes,
        'Nulls'         : dataframe.isnull().sum(),
        'Unique Values' : dataframe.nunique(),
        'Completion %'  : ((1 - dataframe.isnull().sum() / len(dataframe)) * 100).round(2)
    })
    display(HTML(f"<h4 style='color:#f5a623; font-family:Trebuchet MS;'>📋 {name} — Quality Snapshot ({dataframe.shape[0]:,} rows × {dataframe.shape[1]} cols):</h4>"))
    display(audit.style
            .background_gradient(cmap='YlOrRd', subset=['Completion %'])
            .format({'Completion %': '{:.1f}%'})
            .set_table_styles([{'selector': 'th', 'props': [('background-color', '#0d1f38'), ('color', '#4fc3f7')]}]))

data_audit(df,            "🌾 yield_df (Master Dataset)")
data_audit(df_pesticides, "🧪 pesticides")
data_audit(df_rainfall,   "🌧️  rainfall")
data_audit(df_temp,       "🌡️  temperature")
```

```python
# ============================================================
# 🛠️ 1.3 — AGRICULTURAL FEATURE ENGINEERING
# ============================================================

# 1. Yield Tier: classify yield into productivity bands
yield_33 = df['hg/ha_yield'].quantile(0.33)
yield_66 = df['hg/ha_yield'].quantile(0.66)
def yield_tier(val):
    if val <= yield_33:   return 'Low Yield'
    elif val <= yield_66: return 'Medium Yield'
    else:                 return 'High Yield'
df['Yield_Tier'] = df['hg/ha_yield'].apply(yield_tier)

# 2. Climate Stress Index: combines rain and temp variability
df['Climate_Stress'] = (
    (df['avg_temp'] / df['avg_temp'].max()) * 0.5 +
    (1 - df['average_rain_fall_mm_per_year'] / df['average_rain_fall_mm_per_year'].max()) * 0.5
).round(4)

# 3. Pesticide Intensity per Yield unit
df['Pesticide_per_Yield'] = (df['pesticides_tonnes'] / (df['hg/ha_yield'] + 1)).round(4)

# 4. Decade grouping
df['Decade'] = (df['Year'] // 10 * 10).astype(str) + 's'

# 5. Log-transformed yield (for ML readiness)
df['Log_Yield'] = np.log1p(df['hg/ha_yield'])

print(f"✅ Feature Engineering Complete. Master Dataset Shape: {df.shape}")
print(f"\n📊 New Features Added:")
print(f"  • Yield_Tier          — Low / Medium / High productivity band")
print(f"  • Climate_Stress      — 0 to 1 composite stress score")
print(f"  • Pesticide_per_Yield — chemical input efficiency metric")
print(f"  • Decade              — grouped time period")
print(f"  • Log_Yield           — log-scaled yield for analysis")
df[['Area','Item','Year','hg/ha_yield','Yield_Tier','Climate_Stress']].head(8)
```

# 🎨 Segment 2: Exploratory Storytelling & Crop Archetypes

Visualization is the bridge between numbers and agricultural narrative. In this segment, we unpack the fundamental distributions of global crop production — **which crops dominate, how yields have trended over time, and which productivity archetypes exist**.

**Key Explorations:**
1. **Crop Yield Distribution** — which crops produce the most per hectare?
2. **Yield Trend Over Time** — is global agricultural productivity improving?
3. **Yield Tier Breakdown** — the High / Medium / Low performers.

```python
# ============================================================
# 🌾 2.1 — CROP ARCHETYPES: AVERAGE YIELD PER CROP
# ============================================================

crop_avg = df.groupby('Item')['hg/ha_yield'].mean().sort_values(ascending=True).reset_index()
crop_avg.columns = ['Crop', 'Avg_Yield']

fig = go.Figure()

fig.add_trace(go.Bar(
    x=crop_avg['Avg_Yield'],
    y=crop_avg['Crop'],
    orientation='h',
    marker=dict(
        color=crop_avg['Avg_Yield'],
        colorscale='YlOrRd',
        showscale=True,
        colorbar=dict(title="Avg Yield<br>(hg/ha)", tickfont=dict(color='white'))
    ),
    text=crop_avg['Avg_Yield'].round(0).astype(int),
    textposition='outside',
    textfont=dict(color='white', size=12)
))

fig.update_layout(
    title=dict(text="<b>🌾 Crop Productivity Ranking — Average Yield (hg/ha)</b>",
               font=dict(size=20, color=C_GOLD)),
    xaxis=dict(title="Average Yield (hg/ha)", color='white', gridcolor='#1e3a5f'),
    yaxis=dict(color='white', gridcolor='#1e3a5f'),
    paper_bgcolor=C_DARK,
    plot_bgcolor='#0d1f38',
    font=dict(color='white'),
    height=500
)

fig.show(renderer="iframe")
```

```python
# ============================================================
# 📈 2.2 — GLOBAL YIELD TREND: 1990–2013
# ============================================================

yearly_crop = df.groupby(['Year', 'Item'])['hg/ha_yield'].mean().reset_index()

fig = go.Figure()

for crop, color in zip(df['Item'].unique(), PALETTE):
    subset = yearly_crop[yearly_crop['Item'] == crop]
    fig.add_trace(go.Scatter(
        x=subset['Year'],
        y=subset['hg/ha_yield'],
        name=crop,
        mode='lines',
        line=dict(color=color, width=2),
        opacity=0.85
    ))

fig.update_layout(
    title=dict(text="<b>📈 24-Year Yield Trend: All 10 Crops (1990–2013)</b>",
               font=dict(size=20, color=C_GOLD)),
    xaxis=dict(title="Year", color='white', gridcolor='#1e3a5f', dtick=2),
    yaxis=dict(title="Avg Yield (hg/ha)", color='white', gridcolor='#1e3a5f'),
    paper_bgcolor=C_DARK,
    plot_bgcolor='#0d1f38',
    font=dict(color='white'),
    legend=dict(bgcolor='rgba(0,0,0,0)', font=dict(size=11)),
    height=520
)

fig.show(renderer="iframe")
```

```python
# ============================================================
# 🍩 2.3 — YIELD TIER BREAKDOWN & DISTRIBUTION
# ============================================================

tier_counts = df['Yield_Tier'].value_counts().reset_index()
tier_counts.columns = ['Tier', 'Count']
tier_colors = [C_GREEN, C_GOLD, C_RED]

fig = make_subplots(
    rows=1, cols=2,
    specs=[[{'type': 'pie'}, {'type': 'xy'}]],
    subplot_titles=("Yield Tier Distribution", "Yield Histogram by Tier")
)

fig.add_trace(
    go.Pie(labels=tier_counts['Tier'], values=tier_counts['Count'],
           hole=0.55,
           marker=dict(colors=tier_colors),
           textfont=dict(size=13)),
    row=1, col=1
)

for tier, color in zip(['High Yield', 'Medium Yield', 'Low Yield'],
                       [C_GREEN, C_GOLD, C_RED]):
    subset = df[df['Yield_Tier'] == tier]['hg/ha_yield']
    fig.add_trace(
        go.Histogram(x=subset, name=tier, marker_color=color,
                     opacity=0.7, nbinsx=40),
        row=1, col=2
    )

fig.update_layout(
    title=dict(text="<b>🍩 Yield Tier Profile: Who's Producing at Scale?</b>",
               font=dict(size=18, color=C_GOLD)),
    paper_bgcolor=C_DARK,
    plot_bgcolor='#0d1f38',
    font=dict(color='white'),
    height=440,
    barmode='overlay',
    legend=dict(bgcolor='rgba(0,0,0,0)')
)
fig.update_xaxes(color='white', gridcolor='#1e3a5f')
fig.update_yaxes(color='white', gridcolor='#1e3a5f')

fig.show(renderer="iframe")
```

# 🌍 Segment 3: Country & Geographic Deep Dive

The world is not equal when it comes to agricultural productivity. In this segment, we map the **global yield landscape** — identifying which countries lead, which lag, and what the regional story looks like for each crop.

**Key Explorations:**
1. **Top 15 & Bottom 15 Countries** — the productivity gap.
2. **Country × Crop Heatmap** — who grows what, and how well?
3. **Yield by Decade × Country** — have countries improved over time?

```python
# ============================================================
# 🏆 3.1 — COUNTRY PRODUCTIVITY: TOP 15 vs BOTTOM 15
# ============================================================

country_avg = df.groupby('Area')['hg/ha_yield'].mean().reset_index()
country_avg.columns = ['Country', 'Avg_Yield']

top15    = country_avg.nlargest(15, 'Avg_Yield')
bottom15 = country_avg.nsmallest(15, 'Avg_Yield')

fig = make_subplots(rows=1, cols=2,
                    subplot_titles=("🏆 Top 15 Producers", "⚠️ Bottom 15 Producers"))

fig.add_trace(
    go.Bar(x=top15['Avg_Yield'], y=top15['Country'], orientation='h',
           marker=dict(color=C_GREEN,
                       line=dict(color=C_GOLD, width=0.5)),
           text=top15['Avg_Yield'].round(0).astype(int),
           textposition='outside', textfont=dict(color='white', size=10)),
    row=1, col=1
)

fig.add_trace(
    go.Bar(x=bottom15['Avg_Yield'], y=bottom15['Country'], orientation='h',
           marker=dict(color=C_RED,
                       line=dict(color=C_GOLD, width=0.5)),
           text=bottom15['Avg_Yield'].round(0).astype(int),
           textposition='outside', textfont=dict(color='white', size=10)),
    row=1, col=2
)

fig.update_layout(
    title=dict(text="<b>🌍 Global Yield Gap: Top 15 vs Bottom 15 Countries</b>",
               font=dict(size=18, color=C_GOLD)),
    paper_bgcolor=C_DARK,
    plot_bgcolor='#0d1f38',
    font=dict(color='white'),
    height=540,
    showlegend=False
)
fig.update_xaxes(color='white', gridcolor='#1e3a5f')
fig.update_yaxes(color='white', gridcolor='#1e3a5f', autorange='reversed')

fig.show(renderer="iframe")
```

```python
# ============================================================
# 🔥 3.2 — COUNTRY × CROP HEATMAP (TOP 20 COUNTRIES)
# ============================================================

top20_countries = df.groupby('Area')['hg/ha_yield'].mean().nlargest(20).index.tolist()
heatmap_df = df[df['Area'].isin(top20_countries)]
pivot = heatmap_df.groupby(['Area', 'Item'])['hg/ha_yield'].mean().unstack(fill_value=0)

fig = go.Figure(data=go.Heatmap(
    z=pivot.values,
    x=pivot.columns.tolist(),
    y=pivot.index.tolist(),
    colorscale='YlOrRd',
    colorbar=dict(title="Avg Yield<br>(hg/ha)", tickfont=dict(color='white')),
    hoverongaps=False
))

fig.update_layout(
    title=dict(text="<b>🗺️ Crop Yield Heatmap: Top 20 Countries × All 10 Crops</b>",
               font=dict(size=18, color=C_GOLD)),
    xaxis=dict(title="Crop", color='white', tickangle=-30),
    yaxis=dict(title="Country", color='white'),
    paper_bgcolor=C_DARK,
    font=dict(color='white'),
    height=560
)

fig.show(renderer="iframe")
```

```python
# ============================================================
# 🗺️ 3.3 — WORLD MAP: AVERAGE YIELD BY COUNTRY
# ============================================================

country_map = df.groupby('Area')['hg/ha_yield'].mean().reset_index()
country_map.columns = ['Country', 'Avg_Yield']

fig = px.choropleth(
    country_map,
    locations='Country',
    locationmode='country names',
    color='Avg_Yield',
    color_continuous_scale='YlOrRd',
    title="<b>🌍 Global Crop Yield Map (Average hg/ha, 1990–2013)</b>",
    labels={'Avg_Yield': 'Avg Yield (hg/ha)'}
)

fig.update_layout(
    paper_bgcolor=C_DARK,
    geo=dict(bgcolor=C_DARK, showframe=False,
             showcoastlines=True, coastlinecolor='#1e3a5f',
             showland=True, landcolor='#0d1f38',
             showocean=True, oceancolor='#071020'),
    font=dict(color='white'),
    title_font=dict(size=18, color=C_GOLD),
    coloraxis_colorbar=dict(tickfont=dict(color='white'), title=dict(font=dict(color='white'))),
    height=500
)

fig.show(renderer="iframe")
```

# ☁️ Segment 4: Climate Drivers — Rain & Temperature

Agriculture lives and dies by the weather. In this segment, we interrogate the relationship between **rainfall, temperature**, and crop yields — uncovering the climate fingerprint on global food production.

**Key Explorations:**
1. **Rainfall vs. Yield** — does more rain mean more food?
2. **Temperature vs. Yield** — finding the crop thermal sweet spot.
3. **Climate Stress Index Distribution** — which crops face the highest environmental pressure?

```python
# ============================================================
# 🌧️ 4.1 — RAINFALL vs YIELD: SCATTER INTELLIGENCE
# ============================================================

fig = go.Figure()

for crop, color in zip(df['Item'].unique(), PALETTE):
    subset = df[df['Item'] == crop]
    fig.add_trace(go.Scatter(
        x=subset['average_rain_fall_mm_per_year'],
        y=subset['hg/ha_yield'],
        mode='markers',
        name=crop,
        marker=dict(color=color, size=5, opacity=0.6,
                    line=dict(color='white', width=0.2)),
        hovertemplate="<b>%{text}</b><br>Rain: %{x} mm<br>Yield: %{y} hg/ha<extra></extra>",
        text=subset['Area']
    ))

fig.update_layout(
    title=dict(text="<b>🌧️ Rainfall vs. Crop Yield — The Hydration Intelligence Map</b>",
               font=dict(size=18, color=C_GOLD)),
    xaxis=dict(title="Average Rainfall (mm/year)", color='white', gridcolor='#1e3a5f'),
    yaxis=dict(title="Yield (hg/ha)", color='white', gridcolor='#1e3a5f'),
    paper_bgcolor=C_DARK,
    plot_bgcolor='#0d1f38',
    font=dict(color='white'),
    legend=dict(bgcolor='rgba(0,0,0,0)', font=dict(size=10)),
    height=520
)

fig.show(renderer="iframe")
```

```python
# ============================================================
# 🌡️ 4.2 — TEMPERATURE BANDS vs YIELD: VIOLIN + BOX
# ============================================================

# Create temperature bands
df['Temp_Band'] = pd.cut(df['avg_temp'],
                          bins=[0, 10, 15, 20, 25, 35],
                          labels=['<10°C', '10–15°C', '15–20°C', '20–25°C', '>25°C'])

fig = go.Figure()

for band, color in zip(['<10°C', '10–15°C', '15–20°C', '20–25°C', '>25°C'],
                       [C_BLUE, C_GREEN, C_GOLD, '#ffa726', C_RED]):
    subset = df[df['Temp_Band'] == band]['hg/ha_yield']
    fig.add_trace(go.Violin(
        y=subset,
        name=band,
        box_visible=True,
        meanline_visible=True,
        fillcolor=color,
        opacity=0.72,
        line_color='white'
    ))

fig.update_layout(
    title=dict(text="<b>🌡️ Temperature Bands vs. Crop Yield — The Thermal Sweet Spot</b>",
               font=dict(size=18, color=C_GOLD)),
    yaxis=dict(title="Yield (hg/ha)", color='white', gridcolor='#1e3a5f'),
    xaxis=dict(title="Average Temperature Band", color='white'),
    paper_bgcolor=C_DARK,
    plot_bgcolor='#0d1f38',
    font=dict(color='white'),
    violinmode='group',
    height=500
)

fig.show(renderer="iframe")
```

```python
# ============================================================
# 💨 4.3 — CLIMATE STRESS INDEX PER CROP
# ============================================================

stress_crop = df.groupby('Item')['Climate_Stress'].mean().sort_values(ascending=False).reset_index()

fig = go.Figure()

fig.add_trace(go.Bar(
    x=stress_crop['Item'],
    y=stress_crop['Climate_Stress'],
    marker=dict(
        color=stress_crop['Climate_Stress'],
        colorscale=[[0, C_GREEN], [0.5, C_GOLD], [1.0, C_RED]],
        showscale=True,
        colorbar=dict(title="Stress Score", tickfont=dict(color='white'))
    ),
    text=stress_crop['Climate_Stress'].round(3),
    textposition='outside',
    textfont=dict(color='white', size=12)
))

fig.update_layout(
    title=dict(text="<b>💨 Climate Stress Index Per Crop (Higher = More Pressure)</b>",
               font=dict(size=18, color=C_GOLD)),
    xaxis=dict(title="Crop", color='white', gridcolor='#1e3a5f', tickangle=-20),
    yaxis=dict(title="Avg Climate Stress Score", color='white', gridcolor='#1e3a5f'),
    paper_bgcolor=C_DARK,
    plot_bgcolor='#0d1f38',
    font=dict(color='white'),
    height=460
)

fig.show(renderer="iframe")
```

# 🧪 Segment 5: Pesticides — The Chemical Edge

Pesticides represent the most human-controlled variable in agriculture. In this segment, we investigate whether chemical inputs translate into yield improvements — or whether they signal desperation in low-productivity regions.

**Key Explorations:**
1. **Pesticide Usage by Country (Top 15)** — who uses the most?
2. **Pesticide Tonnes vs. Yield** — does more chemicals = more food?
3. **Pesticide Trend Over Time** — how has global usage evolved?

```python
# ============================================================
# 🧪 5.1 — TOP 15 PESTICIDE-CONSUMING COUNTRIES
# ============================================================

pest_country = df.groupby('Area')['pesticides_tonnes'].mean().nlargest(15).reset_index()
pest_country.columns = ['Country', 'Avg_Pesticides_Tonnes']

fig = go.Figure()

fig.add_trace(go.Bar(
    x=pest_country['Avg_Pesticides_Tonnes'],
    y=pest_country['Country'],
    orientation='h',
    marker=dict(
        color=pest_country['Avg_Pesticides_Tonnes'],
        colorscale='Oranges',
        showscale=True,
        colorbar=dict(title="Avg Tonnes", tickfont=dict(color='white'))
    ),
    text=pest_country['Avg_Pesticides_Tonnes'].round(0).astype(int),
    textposition='outside',
    textfont=dict(color='white', size=11)
))

fig.update_layout(
    title=dict(text="<b>🧪 Top 15 Countries by Pesticide Consumption (Avg Tonnes/Year)</b>",
               font=dict(size=18, color=C_GOLD)),
    xaxis=dict(title="Avg Pesticides (tonnes)", color='white', gridcolor='#1e3a5f'),
    yaxis=dict(color='white', gridcolor='#1e3a5f', autorange='reversed'),
    paper_bgcolor=C_DARK,
    plot_bgcolor='#0d1f38',
    font=dict(color='white'),
    height=500
)

fig.show(renderer="iframe")
```

```python
# ============================================================
# 🔬 5.2 — PESTICIDE TONNES vs YIELD: THE CHEMICAL TRUTH
# ============================================================

fig = go.Figure()

for tier, color in zip(['High Yield', 'Medium Yield', 'Low Yield'],
                       [C_GREEN, C_GOLD, C_RED]):
    subset = df[df['Yield_Tier'] == tier]
    fig.add_trace(go.Scatter(
        x=subset['pesticides_tonnes'],
        y=subset['hg/ha_yield'],
        mode='markers',
        name=tier,
        marker=dict(color=color, size=6, opacity=0.65,
                    line=dict(color='white', width=0.2)),
        hovertemplate="<b>%{text}</b><br>Pesticides: %{x:,.0f} T<br>Yield: %{y:,.0f}<extra></extra>",
        text=subset['Area']
    ))

fig.update_layout(
    title=dict(text="<b>🔬 Pesticide Usage vs. Crop Yield (coloured by Yield Tier)</b>",
               font=dict(size=18, color=C_GOLD)),
    xaxis=dict(title="Pesticides (tonnes)", color='white', gridcolor='#1e3a5f', type='log'),
    yaxis=dict(title="Yield (hg/ha)", color='white', gridcolor='#1e3a5f'),
    paper_bgcolor=C_DARK,
    plot_bgcolor='#0d1f38',
    font=dict(color='white'),
    legend=dict(bgcolor='rgba(0,0,0,0)'),
    height=510
)

fig.show(renderer="iframe")
```

```python
# ============================================================
# 📅 5.3 — PESTICIDE USAGE TREND: 1990–2013
# ============================================================

pest_year = df.groupby('Year')['pesticides_tonnes'].mean().reset_index()

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=pest_year['Year'],
    y=pest_year['pesticides_tonnes'],
    mode='lines+markers',
    name='Avg Pesticide Use',
    line=dict(color=C_GOLD, width=3),
    marker=dict(size=8, color=C_RED, line=dict(color='white', width=1)),
    fill='tozeroy',
    fillcolor='rgba(245,166,35,0.15)'
))

fig.update_layout(
    title=dict(text="<b>📅 Global Pesticide Trend (1990–2013) — Rising Chemical Dependency?</b>",
               font=dict(size=18, color=C_GOLD)),
    xaxis=dict(title="Year", color='white', gridcolor='#1e3a5f', dtick=2),
    yaxis=dict(title="Avg Pesticides (tonnes)", color='white', gridcolor='#1e3a5f'),
    paper_bgcolor=C_DARK,
    plot_bgcolor='#0d1f38',
    font=dict(color='white'),
    height=440
)

fig.show(renderer="iframe")
```

# 📈 Segment 6: Correlation & Pattern Discovery

Numbers tell stories — but correlation maps reveal **hidden agricultural narratives**. In this segment, we use statistical correlation, sankey flows, and sunburst charts to understand the invisible threads connecting climate, chemicals, and crop outcomes.

**Key Explorations:**
1. **Full Correlation Matrix** — the numerical relationship web.
2. **Sunburst: Country → Crop → Yield Tier** — the full hierarchy.
3. **Sankey Flow: Crop → Yield Tier** — tracking productivity flow.

```python
# ============================================================
# 🔗 6.1 — FULL CORRELATION MATRIX
# ============================================================

num_cols = ['hg/ha_yield', 'average_rain_fall_mm_per_year',
            'pesticides_tonnes', 'avg_temp',
            'Climate_Stress', 'Pesticide_per_Yield', 'Log_Yield']

corr = df[num_cols].corr()

fig = go.Figure(data=go.Heatmap(
    z=corr.values,
    x=corr.columns.tolist(),
    y=corr.index.tolist(),
    colorscale='RdYlGn',
    zmin=-1, zmax=1,
    text=corr.round(2).values,
    texttemplate="%{text}",
    textfont=dict(size=12),
    colorbar=dict(title="Correlation", tickfont=dict(color='white'))
))

fig.update_layout(
    title=dict(text="<b>🔗 Correlation Matrix: The Agricultural Intelligence Web</b>",
               font=dict(size=18, color=C_GOLD)),
    xaxis=dict(color='white', tickangle=-30),
    yaxis=dict(color='white'),
    paper_bgcolor=C_DARK,
    font=dict(color='white'),
    height=500
)

fig.show(renderer="iframe")
```

```python
# ============================================================
# 🌞 6.2 — SUNBURST: CROP → YIELD TIER HIERARCHY
# ============================================================

sunburst_df = df.groupby(['Item', 'Yield_Tier']).size().reset_index(name='Count')

fig = px.sunburst(
    sunburst_df,
    path=['Item', 'Yield_Tier'],
    values='Count',
    color='Item',
    color_discrete_sequence=PALETTE,
    title="<b>🌞 Crop → Yield Tier Hierarchy: The Full Production Picture</b>"
)

fig.update_traces(textfont=dict(size=13))
fig.update_layout(
    paper_bgcolor=C_DARK,
    font=dict(color='white', size=13),
    title_font=dict(size=18, color=C_GOLD),
    height=540
)

fig.show(renderer="iframe")
```

```python
# ============================================================
# 🌊 6.3 — SANKEY FLOW: CROP → YIELD TIER
# ============================================================

crops    = df['Item'].unique().tolist()
tiers    = df['Yield_Tier'].unique().tolist()
all_lbls = crops + tiers
lbl_idx  = {l: i for i, l in enumerate(all_lbls)}

sankey_data = df.groupby(['Item', 'Yield_Tier']).size().reset_index(name='Count')

node_colors = PALETTE[:len(crops)] + [C_GREEN, C_GOLD, C_RED]

fig = go.Figure(data=[go.Sankey(
    node=dict(
        pad=20, thickness=28,
        line=dict(color="white", width=0.5),
        label=all_lbls,
        color=node_colors[:len(all_lbls)],
        hovertemplate='%{label}: %{value}<extra></extra>'
    ),
    link=dict(
        source=[lbl_idx[r['Item']]       for _, r in sankey_data.iterrows()],
        target=[lbl_idx[r['Yield_Tier']] for _, r in sankey_data.iterrows()],
        value=sankey_data['Count'].tolist(),
        color='rgba(245,166,35,0.25)'
    )
)])

fig.update_layout(
    title=dict(text="<b>🌊 Sankey Flow: Crop Type → Yield Tier Classification</b>",
               font=dict(size=18, color=C_GOLD)),
    paper_bgcolor=C_DARK,
    font=dict(color='white', size=13),
    height=520
)

fig.show(renderer="iframe")
```

```python
# ============================================================
# 🕰️ 6.4 — DECADE-WISE YIELD PROGRESS PER CROP
# ============================================================

decade_crop = df.groupby(['Decade', 'Item'])['hg/ha_yield'].mean().reset_index()

fig = px.bar(
    decade_crop,
    x='Item', y='hg/ha_yield', color='Decade',
    barmode='group',
    color_discrete_sequence=[C_BLUE, C_GOLD, C_GREEN],
    title="<b>🕰️ Decade-wise Yield Progress: Are We Farming Better?</b>",
    labels={'hg/ha_yield': 'Avg Yield (hg/ha)', 'Item': 'Crop'}
)

fig.update_layout(
    paper_bgcolor=C_DARK,
    plot_bgcolor='#0d1f38',
    font=dict(color='white'),
    title_font=dict(size=18, color=C_GOLD),
    xaxis=dict(color='white', gridcolor='#1e3a5f', tickangle=-20),
    yaxis=dict(color='white', gridcolor='#1e3a5f'),
    legend=dict(bgcolor='rgba(0,0,0,0)'),
    height=480
)

fig.show(renderer="iframe")
```
